```python
#import openpyxl
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
# Working directory setup

os.chdir('C:/Users/marco/Dropbox (MPD)/Analytics Argentina/Personal Files/MM/random_downloads/Python/Inflation')
codes=pd.read_csv('codebook.csv')
```

```python
columns=codes.columns[0].replace('       ',' ').replace('      ',' ').replace('   ',' ')
columns=re.split(' ',columns)
col_n=len(columns)

codes=codes.iloc[:,:col_n]
codes.columns=columns

codes['item_code']=codes['item_code'].astype('str')
codes['item_name']=codes['item_name'].astype('str')
codes['display_level']=codes['display_level'].astype('str')
codes['selectable']=codes['selectable'].astype('str')
codes['sort_sequence']=codes['sort_sequence'].astype('str')
```

```python
all_splits={}
for index, row in codes.iterrows():
    no_whitespace=codes.at[index,'item_code'].replace('         ','#').replace('        ','#').replace('       ','#').replace('      ','#').replace('     ','#').replace('    ','#').replace('   ','#').replace('  ','#')
    splits= no_whitespace.split('#')
    all_splits[index]=splits
```

```python
count_1=0
count_2=0
count_3=0
count_4=0
count_5=0
count_6=0
count_7=0
count_8=0
count_9=0
count_10=0

length3=[]
length4=[]
length5=[]

for index, s in all_splits.items():
    if len(s)==1:
        count_1+=1
    elif len(s)==2:
        count_2+=1
    elif len(s)==3:
        count_3+=1
        length3.append(index)
    elif len(s)==4:
        count_4+=1
        length4.append(index)
    elif len(s)==5:
        count_5+=1
        length5.append(index)
    elif len(s)==6:
        count_6+=1
    elif len(s)==7:
        count_7+=1
    elif len(s)==8:
        count_8+=1
    elif len(s)==9:
        count_9+=1
    elif len(s)==10:
        count_10+=1
    else:
        print("unexpected ocurrence")
```
